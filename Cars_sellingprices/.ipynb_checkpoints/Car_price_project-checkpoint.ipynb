```python
# Define a list of categorical columns to be one-hot encoded
categorical_columns = ['year','fuel', 'seller_type', 'transmission', 'owner', 'seats']

# Loop through each categorical field for one-hot encoding
for field in categorical_columns:
    # Create dummy variables for the current field
    dummies = pd.get_dummies(data[field], prefix=field, drop_first=False)
    # Concatenate the dummy variables with the original DataFrame along columns
    data = pd.concat([data, dummies], axis=1)
# Drop the original categorical fields that have been one-hot encoded
data = data.drop(categorical_columns, axis=1)
# Display the first few rows of the modified DataFrame
data.head()
```

```python
# Define the features and target variable

#unique_permno = data['selling_price'].unique()[1:]
#feautures = data[data['selling_price'].isin(unique_permno)]
target = 'selling_price'
features = np.setdiff1d(data.columns, np.array(["selling_price", "name"]))[1:]
features
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is your dataset with columns: 'year', 'fuel', 'km_driven', and 'selling_price'

# Split the data into a train set and a validation set
train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)

# Initialize lists to store MSE values for training and validation sets
mse_train = []
mse_val = []

# Vary the minimum size of terminal nodes
min_samples_leaf_values = list(range(10, 101, 10))

# Vary the max_features hyperparameter
max_features_values = [1, 2, 3]  # Experiment with different values

for min_samples_leaf in min_samples_leaf_values:
    # Create and train the Random Forest model with varying min_samples_leaf
    regr = RandomForestRegressor(n_estimators=20, min_samples_leaf=min_samples_leaf, max_features=3, random_state=0)  # Use max_features=3
    regr.fit(train_data[features], train_data[target])

    # Predict on the training set
    y_train_pred = regr.predict(train_data[features])
    mse_train.append(mean_squared_error(train_data[target], y_train_pred))

    # Predict on the validation set
    y_val_pred = regr.predict(val_data[features])
    mse_val.append(mean_squared_error(val_data[target], y_val_pred))

# Plot the MSE values as a function of min_samples_leaf
plt.figure(figsize=(8, 6))
plt.plot(min_samples_leaf_values, mse_train, marker='o', label='Training MSE')
plt.plot(min_samples_leaf_values, mse_val, marker='o', label='Validation MSE')
plt.xlabel('Min Samples Leaf')
plt.ylabel('Mean Squared Error (MSE)')
plt.legend()
plt.title('MSE vs. Min Samples Leaf')
plt.grid(True)
plt.show()

# Find the best min_samples_leaf value
best_min_samples_leaf = min_samples_leaf_values[np.argmin(mse_val)]
print('Best min_samples_leaf:', best_min_samples_leaf)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is your dataset with columns: 'year', 'fuel', 'km_driven', and 'selling_price'

# Split the data into a train set and a validation set
train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)

# Initialize lists to store MSE values for training and validation sets
mse_train = []
mse_val = []

# Vary the minimum size of terminal nodes
min_samples_leaf_values = list(range(10, 101, 10))

# Vary the max_features hyperparameter
max_features_values = [1, 2, 3]  # Experiment with different values

for min_samples_leaf in min_samples_leaf_values:
    for max_features in max_features_values:
        # Create and train the Random Forest model with varying min_samples_leaf and max_features
        regr = RandomForestRegressor(n_estimators=20, min_samples_leaf=min_samples_leaf, max_features=max_features, random_state=0)
        regr.fit(train_data[features], train_data[target])

        # Predict on the training set
        y_train_pred = regr.predict(train_data[features])
        mse_train.append(mean_squared_error(train_data[target], y_train_pred))

        # Predict on the validation set
        y_val_pred = regr.predict(val_data[features])
        mse_val.append(mean_squared_error(val_data[target], y_val_pred))

# Reshape the MSE values for plotting
mse_train = np.array(mse_train).reshape(len(min_samples_leaf_values), len(max_features_values))
mse_val = np.array(mse_val).reshape(len(min_samples_leaf_values), len(max_features_values))

# Create a heatmap of MSE values as a function of min_samples_leaf and max_features
plt.figure(figsize=(10, 6))
plt.imshow(mse_val, cmap='hot', interpolation='nearest', aspect='auto', origin='lower')
plt.colorbar(label='Validation MSE')
plt.xticks(np.arange(len(max_features_values)), max_features_values)
plt.yticks(np.arange(len(min_samples_leaf_values)), min_samples_leaf_values)
plt.xlabel('Max Features')
plt.ylabel('Min Samples Leaf')
plt.title('Validation MSE Heatmap')
plt.show()

# Find the best min_samples_leaf and max_features values
best_indices = np.unravel_index(np.argmin(mse_val), mse_val.shape)
best_min_samples_leaf = min_samples_leaf_values[best_indices[0]]
best_max_features = max_features_values[best_indices[1]]
print('Best min_samples_leaf:', best_min_samples_leaf)
print('Best max_features:', best_max_features)
```
